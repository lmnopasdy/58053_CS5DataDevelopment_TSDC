```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (30,12)
x = df['Name']
y = df['HP']

plt.xlabel('Name', fontsize=22)
plt.ylabel('HP', fontsize=22)
plt.bar(x, y)
df.sort_values('Name',inplace=True)
##plt.title("HP per Heroes", fontsize = 30)
plt.show
```

```python
import seaborn as sns
sns.set_style('darkgrid')
ax = sns.barplot(x = 'Element',y = 'HP', data = df, ci = None);
plt.xlabel('Element', fontsize=25, weight = 'bold')
plt.ylabel('HP', fontsize=25, weight = 'bold' )
plt.tick_params(labelsize=16) 
df.sort_values('Element',inplace=True)
```

```python
df.groupby('Element').HP.mean()
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.barplot(x = 'Element',y = 'HP', data = df, ci = None, estimator = sum)
plt.title("Overall HP per Element", fontsize = 30)
plt.xlabel('Element', fontsize=22)
plt.ylabel('HP', fontsize=22)
plt.tick_params(labelsize=16)
```

```python
df.groupby('Element').HP.sum()
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.barplot(x = 'Element',y = 'Atk', data = df, ci = None, estimator = sum);
plt.title("Overall Attack per Element", fontsize = 30)
plt.xlabel('Element', fontsize=22)
plt.ylabel('Atk', fontsize=22)
plt.tick_params(labelsize=16)
```

```python
df.groupby('Element').Atk.sum()
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.barplot(x = 'Element',y = 'Def', data = df, ci = None, estimator = sum);
plt.title("Overall Defense per Element", fontsize = 30)
plt.xlabel('Element', fontsize=22)
plt.ylabel('Def', fontsize=22)
plt.tick_params(labelsize=16)
```

```python
df.groupby('Element').Def.sum()
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.barplot(x = 'Name',y = 'Atk', data = df, ci = None, order=df.sort_values('Atk',ascending = False).Name);
plt.title("Heroes Attack", fontsize = 30)
plt.xlabel('Name', fontsize=22)
plt.ylabel('Atk', fontsize=22)
plt.tick_params(labelsize=10)
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.barplot(x = 'Name',y = 'HP', data = df, ci = None, order=df.sort_values('HP',ascending = False).Name);
plt.title("Heroes HP", fontsize = 30)
plt.xlabel('Name', fontsize=22)
plt.ylabel('HP', fontsize=22)
plt.tick_params(labelsize=10)
```

```python
import seaborn as sns
sns.set_style('darkgrid')
sns.barplot(x = 'Name',y = 'Def', data = df, ci = None,order=df.sort_values('Def',ascending = False).Name);
plt.title("Heroes Defense", fontsize = 30)
plt.xlabel('Name', fontsize=22)
plt.ylabel('Def', fontsize=22)
plt.tick_params(labelsize=10)
```
